```python
# dicionario que mapeia numeros para letras
nos =  {
0 : 'A',
1 : 'B',
2 : 'C',
3 : 'D',
4 : 'E',
5 : 'F',
6 : 'G',
7 : 'H',
8 : 'I',
9 : 'J',
10: 'K'
}

# lista de lista de vizinhos / os caminhos do grafo ou entre os nos 
nos_visinhos =   [ [1,2],     # visiho do no 0
                   [3,4],     # visinho do no 1
                   [5],   # visiho do no 2
                   [7],       # visiho do no 3
                   [5],       # visiho do no 4
                   [6],        # visiho do no 5
                   [7],        # visiho do no 6
                   [8],       # visiho do no 7
                   [9,10],    # visiho do no 8
                   [10],        # visiho do no 9
                   [],        # visiho do no 10
]
# lista de lista de pesos / custos entre os nos
nos_custos =     [ [1,4],        # custo do no 0 para 1 e 2 respectivamente
                   [2,5],        # custo do no 1 para 3 e 4 respectivamente
                   [2,3],        # custo do no 2 para 5 e 6 respectivamente
                   [6],          # custo do no 3 para 7  
                   [2],           # custo do no 4 para 5
                   [2],           # custo do no 5 para 6
                   [2],           # custo do no 6 para 7
                   [1],          # custo do no 7 para 8
                   [2,0],        # custo do no 8 para 9 e 10 respectivamente
                   [5],           # custo do no 9 para 10
                   [0],           # custo do no 10 para 0
]
# heuristica para o A*
heuristica = {
    0:  9,   # A (Custo real até K é 10. Chutamos 9)
    1:  8,   # B (Custo real até K é 9. Chutamos 8)
    2:  15,  # C (Caminho ruim/sem saída para K. Valor alto para desencorajar)
    3:  6,   # D (Custo real até K é 7. Chutamos 6)
    4:  20,  # E (Sem saída)
    5:  20,  # F (Sem saída)
    6:  20,  # G (Sem saída)
    7:  1,   # H (Custo real até K é 1. Chutamos 1)
    8:  0,   # I (Custo real é 0, pois a aresta I->K tem peso 0)
    9:  5,   # J (Vizinho de I, mas não é o alvo)
    10: 0    # K (Objetivo = distância 0)
}


def A_estrela():
    lista = []
    objetivo = 10  # No K
    inicio = 0     # No A
    lista.append(inicio)
    while len(lista) > 0:
        no = lista.pop(0)
        print(f"Visitando nó {nos[no]}")
        if no == objetivo:
            print("Objetivo alcançado!")
            return
        for vizinho in nos_visinhos[no]:
            if vizinho not in lista:
                lista.append(vizinho)
        print(lista)
        menor =  heuristica[lista[0]] + nos_custos[no][nos_visinhos[no].index(lista[0])]
        valor_menor = lista[0]
        for n in lista[1:]:
            aux = heuristica[n] + nos_custos[no][nos_visinhos[no].index(n)]
            if aux < menor:
                menor = aux
                valor_menor = n
        print(f"Próximo nó a visitar: {nos[valor_menor]} com valor {menor}")
        lista = []
        lista.append(valor_menor)

teste = A_estrela()
```
