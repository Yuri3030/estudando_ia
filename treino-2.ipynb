```python
from typing import Any, Callable, Iterable, Optional, List, Tuple, Dict, Set
import heapq              # fila de prioridade (UCS / A*)
from collections import deque  # fila (BFS)
import math               # inf, etc.
import random
from time import sleep
```

```python
lista = []
tuplas = []
for i in range(4):
    lista.append(random.randint(1,6))
print(lista)
for i, numero in enumerate(lista):
    print(f'O jogador {i} tirou  {numero}')
    tupla = (i, numero)
    tuplas.append(tupla)

print(tuplas)
tuplas_ordenadas = sorted(tuplas, key=lambda y: y[1], reverse=True)
print(tuplas_ordenadas)
```

```python

jogadores= {}

for i in range(4):
    nome = 'jogador ' + str(i)
    jogadores[nome] = random.randint(1,6) 
print(jogadores)

ordenado_valor = dict(sorted(jogadores.items(), key=lambda item: item[1], reverse=True))
print(ordenado_valor)
for chave, valor in ordenado_valor.items():
    sleep(1)
    print(f' O {chave} tirou {valor}  no dado')
```

```python
aluno = {}
aluno['nome'] = str(input('Seu Nome: '))
aluno['media'] = float(input(f'meida de {aluno["nome"]}: '))

if aluno['media'] >= 7:
    aluno['situacao'] = 'Aprovado'
elif 5 <= aluno['media'] < 7:
    aluno['situacao'] = 'Recuperação'
else:
    aluno['situacao'] = 'Reprovado'

print(f'Aluno: {aluno["nome"]}\nMédia: {aluno["media"]}\nSituação: {aluno["situacao"]}')
```
