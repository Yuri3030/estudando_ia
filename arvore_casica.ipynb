```python

nos =  {
0 : 'A',
1 : 'B',
2 : 'C',
3 : 'D',
4 : 'E',
5 : 'F',
6 : 'G',
7 : 'H',
8 : 'I',
9 : 'J',
10: 'K'
}
nos_visinhos =   [ [1,2],    # visiho do no 0
                   [3,4],    # visinho do no 1
                   [5,6,],   # visiho do no 2
                   [7],   # visiho do no 3
                   [],       # visiho do no 4
                   [],       # visiho do no 5
                   [],       # visiho do no 6
                   [8],      # visiho do no 7
                   [9,10],   # visiho do no 8
                   [],       # visiho do no 9
                   [],       # visiho do no 10
]
def buscar_indice_linear(meta):
    for chave, valor in nos.items():
        if valor == meta:
            return chave
    return None # Não encontrado

def busca_em_profundidade (no_inicial,no_objetivo):
    fila = deque()
    
    no = buscar_indice_linear(no_inicial)
    objetivo = buscar_indice_linear(no_objetivo)
    fila.append(no)
    print("fila atual:", fila)
    no = fila.popleft()
    print("visitando nó:",nos[no])
    print("fila atual:", fila)

    while no is not None:
        # esse nó é o objetivo?
        if no == objetivo:
            print("nó objetivo encontrado:",nos[no])
            return True
        else:
              
            for vizinho in reversed(nos_visinhos[no]):
                fila.appendleft(vizinho)
                print("fila atual:", fila)
        no = fila.popleft()
        print("visitando nó:",nos[no])

    print("nó objetivo não encontrado")
    return False


teste = busca_em_profundidade("A","K")
print(teste)
```
