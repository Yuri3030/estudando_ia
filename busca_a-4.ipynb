```python
grafo_nos = {
               'S':{'A':3},
               'A':{'B':1, 'C':2,'D': 3},
               'B':{'E':4},
               'C':{'E':5},
               'D':{'E':1},
               'E':{}
}

heuristica = {
    'S':7,
    'A':6,
    'B':5,
    'C':6,
    'D':2,
    'E':0
}
no_inicial = 'S'
no_objetivo = 'E'
aberto = {no_inicial}
g = {no_inicial: 0}


no_atual = no_inicial
while aberto:
    if no_atual == no_objetivo:
        print("Objetivo alcançado:", no_atual)
        print("Custo total:", g[no_atual])
        break
    print("nos abertos:", aberto)
    aberto.remove(no_atual)
    print("no atual:", no_atual)
    vizinhos_no =  grafo_nos[no_atual]
    print("vizinhos do no atual:", vizinhos_no)
    for vizinho, custo in vizinhos_no.items():
        print("vizinho:", vizinho)
        aberto.add(vizinho)
        g[vizinho] = g[no_atual] + custo
    print("g:", g)
    #no_atual = min(aberto, key=lambda no: g[no] + heuristica[no])
    Func = {}
    for no in aberto:
        Func[no] = g[no] + heuristica[no]
    aux = 0
    for no, f in Func.items():
        if aux == 0:
            no_atual = no
            aux = f
        elif f < aux:
            no_atual = no
            aux = f
    print("Func:", Func)
    print("no atual:", no_atual)





```
