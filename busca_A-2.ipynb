```python
nomes_dos_Nos = { 0: "S",
                  1: "A", 
                  2: "B",
                  3: "C", 
                  4: "D",
                  5: "E"
}
nos_e_Seus_vizinhos = { 0: [1],
                        1: [2,3],
                        2: [4,5],
                        3: [5],
                        4: [5],
                        5: []
}
custos_das_Arestas_entre_nos = {  (0,1): 1,
                                  (1,2): 1,
                                  (1,3): 2,
                                  (2,4): 3,
                                  (2,5): 4,
                                  (3,5): 5,
                                  (4,5): 2
}
heuristicas_dos_Nos = { 0: 5,
                        1: 4,
                        2: 3,
                        3: 4,
                        4: 2,
                        5: 0
}

no_de_inicio = 0
no_objetivo = 5

def estrela (no_objetivo, no_de_inicio):
    # lista para procurar os nos
    lista_procura = []
    lista_procura.append(no_de_inicio)
    # lista para armazenar os nos explorados
    lista_nos_explorados = []
    lista_vizinhos = []
    while len(lista_procura) > 0:
        no_atual = lista_procura.pop(0)
        print("nó atual:",no_atual)
        if no_atual == no_objetivo:
            print("Objetivo alcançado!")
        else:
            # adiciona o nó atual na lista de nós explorados
            lista_nos_explorados.append(no_atual)
            print("lista de nós explorados:", lista_nos_explorados)
            for vizinho in nos_e_Seus_vizinhos[no_atual]:
                if vizinho not in lista_nos_explorados and vizinho not in lista_procura:
                    lista_vizinhos.append(vizinho)
            print("lista de vizinhos:", lista_vizinhos)
            # chama a função heurística para escolher o próximo nó
            proximo_no = funcao_herustica(no_atual, lista_vizinhos)
            lista_procura.append(proximo_no)
            print("lista de procura:", lista_procura)
            lista_vizinhos.clear()
        
def funcao_herustica( no_atual: int,lista_vizinhos :list[int] )-> list[int] :
    dicionario_heuristica = {}
    for vizinho in lista_vizinhos:
        dicionario_heuristica[vizinho] = heuristicas_dos_Nos[vizinho] + custos_das_Arestas_entre_nos[(no_atual,vizinho)]
    dicionario_ordenado = sorted(dicionario_heuristica.items(), key=lambda x: x[1])
    proximo_no = dicionario_ordenado[0][0]
    print("dicionario ordenado", dicionario_ordenado)
    print("esse é o proximo no" ,proximo_no)
    return proximo_no

estrela(no_objetivo, no_de_inicio)
```
